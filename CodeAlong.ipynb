```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = 'data_science_salary_2021 - data_science_salary_2021.csv'
df = pd.read_csv(data)
df.head()
```

```python
# Group the DataFrame by 'job_title_sim' and calculate the mean 'Avg Salary(K)'
mean_salary_by_job_title = df.groupby('job_title_sim')['Avg Salary(K)'].mean()
print(mean_salary_by_job_title)
```

```python
# Plotting the mean salary by job_title_sim
plt.figure(figsize=(10, 6))
mean_salary_by_job_title.plot(kind='bar', color='lightgreen')
plt.title('Mean Avg Salary(K) by Job Title Similarity')
plt.xlabel('Job Title Similarity')
plt.ylabel('Mean Avg Salary(K)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
